```python
# 1. Import Library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 2. Load Data
df = pd.read_csv('train.csv')

# 3. Tampilkan 5 baris awal
df.head()

# 4. Info struktur data
df.info()

# 5. Statistik deskriptif
df.describe()

# 6. Cek missing values
df.isnull().sum()

# 7. Visualisasi

# a. Selamat atau tidak
sns.countplot(x='Survived', data=df)
plt.title('Jumlah Penumpang yang Selamat vs Tidak')
plt.show()

# b. Distribusi penumpang berdasarkan jenis kelamin
sns.countplot(x='Sex', data=df)
plt.title('Distribusi Penumpang Berdasarkan Jenis Kelamin')
plt.show()

# c. Kombinasi jenis kelamin & keselamatan
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Keselamatan Berdasarkan Jenis Kelamin')
plt.show()

```
